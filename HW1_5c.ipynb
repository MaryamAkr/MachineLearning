### Question 5- Part C
Use the numpy random.randn to simulate 10000 data points, 5,000 from bean type A and 5,000 from bean type B. You can now pick 1000 values of T using linspace between T = 1 and T = 8. For each of these you can compute the miss classification rate. Make the figure. These should match closely your results for b above.

Knowing that :

The mean weight of bean type A, μA is 5 grams and the standard deviation,
σA, is 2. 

The mean weight of bean type B, muB, is 4 grams and has a standard deviation σB, of 1.4.

```python
import numpy as np
from matplotlib import pyplot as plt

mu_A = 5.0
sigma_A = 2.0
mu_B = 4.0
sigma_B = 1.4
n_sample = 10000
np.random.seed(1234)

bean_A = [sigma_A* np.random.randn() + mu_A for i in range(int(n_sample/2))]
bean_B = [sigma_B* np.random.randn() + mu_B for i in range(int(n_sample/2))]

T = np.linspace(1,8,1000)

from scipy.stats import norm

def error(T):
    return (1 - (0.5*norm.cdf(T,mu_A,sigma_A) + 0.5*(1 - norm.cdf(T,mu_B,sigma_B))))


bayes_error = list(map(error,T))


plt.plot(T,bayes_error,  color='blue',label='bayes_error')
plt.legend()
plt.xlabel('T')
plt.ylabel('Bayes_error')
plt.show()
```
